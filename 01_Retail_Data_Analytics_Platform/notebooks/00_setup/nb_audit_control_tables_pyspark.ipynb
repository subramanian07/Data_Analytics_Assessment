```python
# nb_00_setup_control_tables
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, TimestampType
from pyspark.sql import Row

spark.sql("CREATE DATABASE IF NOT EXISTS ETL")

table_names = ["ETL.pl_control_config_log", "ETL.pl_bad_records_log", "ETL.pl_watermark_log"]
```

```python
# ==================================
# ETL pl_control_config_log TABLE
# ==================================
spark.sql(f"DROP TABLE IF EXISTS {table_names[0]}")

etl_log_schema = StructType([
    StructField("run_id", StringType(), True),
    StructField("pipeline_name", StringType(), True),
    StructField("pipeline_step_name", StringType(), True),
    StructField("source_file_path", StringType(), True),
    StructField("target_table_name", StringType(), True),
    StructField("step_status", StringType(), True),
    StructField("pipeline_status", StringType(), True),
    StructField("record_count", IntegerType(), True),
    StructField("error_message", StringType(), True),
    StructField("step_start_time", TimestampType(), True),
    StructField("step_end_time", TimestampType(), True),
    StructField("pipeline_start_time", TimestampType(), True),
    StructField("pipeline_end_time", TimestampType(), True),
    StructField("created_at", TimestampType(), True),
    StructField("updated_at", TimestampType(), True)
])

df_etl_log = spark.createDataFrame([], etl_log_schema)

df_etl_log.write.format("delta") \
          .mode("overwrite") \
          .saveAsTable(table_names[0])

print(f"{table_names[0]} created successfully")
```

```python
# ================================
# ETL pl_bad_records_log Table
# ================================
spark.sql(f"DROP TABLE IF EXISTS {table_names[1]}")

bad_records_schema = StructType([
    StructField("run_id", StringType(), True),
    StructField("pipeline_name", StringType(), True),
    StructField("pipeline_step_name", StringType(), True),
    StructField("error_data", StringType(), True),
    StructField("error_message", StringType(), True),
    StructField("created_at", TimestampType(), True)
])

df_bad_records_log = spark.createDataFrame([], bad_records_schema)

df_bad_records_log.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(table_names[1])

print(f"{table_names[1]} created successfully")
```

```python
# ==================================
# ETL pl_watermark_log TABLE
# ==================================
spark.sql(f"DROP TABLE IF EXISTS {table_names[2]}")

etl_log_schema = StructType([
    StructField("run_id", StringType(), True),
    StructField("pipeline_name", StringType(), True),
    StructField("pipeline_step_name", StringType(), True),
    StructField("source_name", StringType(), True),
    StructField("watermark_column", StringType(), True),
    StructField("watermark_value", StringType(), True),
    StructField("updated_at", TimestampType(), True),
])

df_etl_log = spark.createDataFrame([], etl_log_schema)

df_etl_log.write.format("delta") \
          .mode("overwrite") \
          .saveAsTable(table_names[2])

print(f"{table_names[2]} created successfully")
```
